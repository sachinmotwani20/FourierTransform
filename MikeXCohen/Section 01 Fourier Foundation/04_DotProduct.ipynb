```
import numpy as np
import matplotlib.pyplot as plt
```

Dot Product of Two Vectors

```
#Two Vectors
v1 = [1, 2, 3]
v2 = [3, 2, 1]

#Compute the Dot Product
dp = sum(np.multiply(v1,v2))

print('The dot product is ', dp)
```

Dot Product of Sine Wave

```
#General Simulation Parameters
srate = 500; #Sampling Rate in Hz
time = np.arange(0.0, 2.0, 1.0/srate) #Time in Seconds

#Sine Wave Parameters
freq1 = 5
freq2 = 5

ampl1 = 2
ampl2 = 2

phas1 = np.pi/2
phas2 = np.pi/2

#Generate the Sine Wave
sinewave1 = ampl1 * np.sin(2*np.pi * freq1 * time + phas1)
sinewave2 = ampl2 * np.sin(2*np.pi * freq2 * time + phas2)

#Compute dot product
dp = np.dot (sinewave1, sinewave2)

#Print Result
print('Dot Product = ', dp)
```

Alter some of these Wave Parameters above, to understand the variation in the Dot Product

Dot product of two Sine Waves with different frequencies is Zero (approx).
[This is only true when the frequencies are in the multiples of 1 or 0.5;

For rest, the output will be non-zero]

Reason: Sampling Rate

```
#General Simulation Parameters
srate = 500; #Sampling Rate in Hz
time = np.arange(0.0, 2.0, 1.0/srate) #Time in Seconds

#Sine Wave Parameters
freq1 = 6
freq2 = 5

ampl1 = 2
ampl2 = 2

phas1 = np.pi/2
phas2 = np.pi/2

#Generate the Sine Wave
sinewave1 = ampl1 * np.sin(2*np.pi * freq1 * time + phas1)
sinewave2 = ampl2 * np.sin(2*np.pi * freq2 * time + phas2)

#Compute dot product
dp = np.dot (sinewave1, sinewave2)

#Print Result
print('Dot Product = ', dp)
```

Dot Product of Signals with Seroies of Sine Waves

```
#Phase of signal 
theta = 0*np.pi/4

#Simulation Parameters
srate = 1000
time  = np.arange (-1.0, 1.0, 1.0/srate)

#Signal
sinew = np.sin(2*np.pi*5*time + theta) #Sine Wave
gauss = np.exp((-time**2)/0.1) #Gaussian
signal = np.multiply(sinew, gauss)  #Morlet Wavelet : It's a sine Wave tapered by a Gaussian

#Vector of Sine Wave Frequencies; to be used for these sine waves
sinefrex = np.arange(2.0, 10.0, 0.5)

#Plot Signal
plt.plot(time, signal)
plt.xlabel('time (sec.)')
plt.ylabel('amplitude (a.u.')
plt.title('Signal')
plt.show()


#Initialize dot products vectors
dps = np.zeros(len(sinefrex))

#Loop over Sine Waves

for fi in range (1, len(dps)):
    #create a new sine wave; Changes with every iteration
    sinew = np.sin( 2 *np.pi*sinefrex[fi]* time)

    #compute dot product
    dps[fi] = np.dot(sinew, signal)/len(time)

# and plot
plt.stem(sinefrex, dps)
plt.xlabel('Sine Wave frequency (Hz)')
plt.ylabel('Dot Product')
plt.title('Dot product with Sine Waves')
plt.show()
```

Note that the dot product gets larger as we approach 5Hz frequency from either side. This is because we have hardcoded the Sine Wave to be of 5Hz. 

The Dot product is just showing the Similarity.

Even though the Dot product isn't directly dependent upon the phase, it changes drastically with the phase.

But this shouldn't happen. The idea of Fourier Transform is to determine the energy present in each frequency of the Signal & in that, phase shouldn't matter. Eg, put phase to be pi/2 should give the answer zero.

```
#Phase of signal 
theta = 2*np.pi/4

#Simulation Parameters
srate = 1000
time  = np.arange (-1.0, 1.0, 1.0/srate)

#Signal
sinew = np.sin(2*np.pi*5*time + theta) #Sine Wave
gauss = np.exp((-time**2)/0.1) #Gaussian
signal = np.multiply(sinew, gauss)  #Morlet Wavelet : It's a sine Wave tapered by a Gaussian

#Vector of Sine Wave Frequencies; to be used for these sine waves
sinefrex = np.arange(2.0, 10.0, 0.5)

#Plot Signal
plt.plot(time, signal)
plt.xlabel('time (sec.)')
plt.ylabel('amplitude (a.u.')
plt.title('Signal')
plt.show()


#Initialize dot products vectors
dps = np.zeros(len(sinefrex))

#Loop over Sine Waves

for fi in range (1, len(dps)):
    #create a new sine wave; Changes with every iteration
    sinew = np.sin( 2 *np.pi*sinefrex[fi]* time)

    #compute dot product
    dps[fi] = np.dot(sinew, signal)/len(time)

# and plot
plt.stem(sinefrex, dps)
plt.xlabel('Sine Wave frequency (Hz)')
plt.ylabel('Dot Product')
plt.title('Dot product with Sine Waves')
plt.show()
```

Notice the y axis; It is of the range of 10^-17. This is because my computer is precise enough to calculate such changes [Numpy uses a definite approx value of pi, that might not be enough to cancel to outputs]. The answer should be zero. The graph post rounding off is shown below.

```
#Phase of signal 
theta = 2*np.pi/4

#Simulation Parameters
srate = 1000
time  = np.arange (-1.0, 1.0, 1.0/srate)

#Signal
sinew = np.sin(2*np.pi*5*time + theta) #Sine Wave
gauss = np.exp((-time**2)/0.1) #Gaussian
signal = np.multiply(sinew, gauss)  #Morlet Wavelet : It's a sine Wave tapered by a Gaussian

#Vector of Sine Wave Frequencies; to be used for these sine waves
sinefrex = np.arange(2.0, 10.0, 0.5)

#Plot Signal
plt.plot(time, signal)
plt.xlabel('time (sec.)')
plt.ylabel('amplitude (a.u.')
plt.title('Signal')
plt.show()


#Initialize dot products vectors
dps = np.zeros(len(sinefrex))

#Loop over Sine Waves

for fi in range (1, len(dps)):
    #create a new sine wave; Changes with every iteration
    sinew = np.sin( 2 *np.pi*sinefrex[fi]* time)

    #compute dot product
    dps[fi] =np.round( np.dot(sinew, signal)/len(time), 16) #Rounding upto 16 decimal places will give us accurate results.

# and plot
plt.stem(sinefrex, dps)
plt.xlabel('Sine Wave frequency (Hz)')
plt.ylabel('Dot Product')
plt.title('Dot product with Sine Waves')
plt.show()
```
