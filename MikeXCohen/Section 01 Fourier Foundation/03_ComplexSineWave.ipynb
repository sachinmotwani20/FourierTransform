#Sine wave & Complex Sine Wave

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```
#Simulation parameters
srate = 500 #Sampling Rate in Hz
time = np.arange(0.0, 2.0, 1.0/srate) #time in seconds

#Sine Wave Parameters
freq = 3         #Frequency in Hz
ampl = 2         #Amplitude in a.u.
phas = np.pi/3   #Phase in Radians

#Generate the Sine Wave
sine_wave = ampl*np.sin(2*np.pi*freq*time + phas )

plt.plot(time, sine_wave, 'k')
plt.grid(True)
plt.xlabel('Time (sec.)')
plt.ylabel('Amplitude (a.u)')
plt.show()
```

You may play around srate for better understanding of its effect

Sine & Cosine Waves are the same but have a 90 degree phase shift

```
#Generating the waves 
sine_wave = ampl * np.sin( 2*np.pi* freq*time + phas )
cos_wave = ampl * np.cos(2*np.pi*freq * time + phas)

plt.plot(time, sine_wave, 'b', label='Sine')
plt.plot(time, cos_wave, 'r', label='Cosine')
plt.xlabel('Time(sec')
plt.ylabel('Amplitude')
plt.title ('A Cosine Wave is a phase shifted Sine Wave')
plt.show()
```

#Complex Sine Wave

```
#Simulation parameters 
srate = 500 #Sampling Rate in Hz
time = np.arange(0.0, 2.0, 1.0/srate) #Time in Seconds

#Sine Wave Parameters
freq = 5         #frequency in Hz
ampl = 2         #amplitude in a.u.
phas = np.pi/3   #Phase in Radians 


#Generate the Sine Wave
csw = ampl * np.exp(1j*(2*np.pi*freq*time+ phas))

#plot the results 
plt.plot(time, np.real(csw), label=('Real'))
plt.plot(time, np.imag(csw), label=('Imag'))
plt.xlabel('Time (sec.)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

# 3D representation of Complex Sine Waves

```
#Mixed Signals.Learn
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.plot(time, np.real(csw), np.imag(csw))
ax.set_xlabel('Time (s)')
ax.set_ylabel('Reak Part')
ax.set_zlabel('Imaginary Part')
ax.set_title('Complex Sine Wave in 3D')
plt.show()
```
