```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(-3, 3, num=50)

plt.plot(x, np.exp(x), label = ' y = e^x')

plt.axis([min(x), max(x), 0, np.exp(x[-1])])
plt.grid(True)
plt.legend()
plt.xlabel('x')
plt.show()
```

[ cos(k),  sin(k) ] is on the unit circle for any real k

```
# 'k' is any real number
k=2/np.pi

#Eutler's notation
euler = np.exp(1j*k)

#plot dot 
plt.plot(np.cos(k), np.sin(k), 'ro') #red, circular

#draw unit cicle for reference 
x = np.linspace(-np.pi, np.pi, num=100)
plt.plot(np.cos(x), np.sin(x))


plt.grid(True)
plt.axis('scaled')
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.show()
```

```
#Euler's Formula with Arbitrary Vector Magnitude

m = 4; #maqgnitude
k = np.pi/3 #phase
compnum = m*np.exp(1j*k);

#extract magnitude & angle
mag=np.abs(compnum)
phas = np.angle(compnum);

plt.polar ([phas, phas], [0, mag])
plt.show()
```

Notice the Concentric Circles.
